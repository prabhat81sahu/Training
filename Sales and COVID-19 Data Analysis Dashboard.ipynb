```python
df = pd.read_csv('/content/COVID-19 Cases and deaths - WHO.csv')
df
```

```python
df['Daily new confirmed cases of COVID-19'] = pd.to_numeric(df['Daily new confirmed cases of COVID-19'], errors='coerce').fillna(0)

country_total_cases = df.groupby('Country')['Daily new confirmed cases of COVID-19'].sum().reset_index()
country_total_cases = country_total_cases.sort_values(by='Daily new confirmed cases of COVID-19', ascending=False)

display(country_total_cases.head(10))
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Daily new confirmed cases of COVID-19', data=country_total_cases.head(10))
plt.title('Top 10 Countries by Total Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df['Daily new confirmed deaths due to COVID-19'] = pd.to_numeric(df['Daily new confirmed deaths due to COVID-19'], errors='coerce').fillna(0)

country_total_deaths = df.groupby('Country')['Daily new confirmed deaths due to COVID-19'].sum().reset_index()
country_total_deaths = country_total_deaths.sort_values(by='Daily new confirmed deaths due to COVID-19', ascending=False)

display(country_total_deaths.head(10))
```

```python
merged_data = pd.merge(country_total_cases, country_total_deaths, on='Country', suffixes=('_cases', '_deaths'))
merged_data['Death Rate (%)'] = (merged_data['Daily new confirmed deaths due to COVID-19'] / merged_data['Daily new confirmed cases of COVID-19']) * 100
merged_data = merged_data.sort_values(by='Death Rate (%)', ascending=False)

display(merged_data.head(10))
```

Finally, I'll visualize the top 10 countries by their death rates using a bar plot to clearly illustrate the comparison.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Death Rate (%)', data=merged_data.head(10))
plt.title('Top 10 Countries by COVID-19 Death Rate')
plt.xlabel('Country')
plt.ylabel('Death Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma vs Virat Kholi Career Camparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='Red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='Blue')

plt.title('Rohit Sharma vs Virat Kholi Career Camparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='Red',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='Blue',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma vs Virat Kholi Career Camparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='Red',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='Blue',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma vs Virat Kholi Career Camparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='Red',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='Blue',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma vs Virat Kholi Career Camparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend()
plt.show()
```

```python
df = pd.read_json('/content/t20_wc_player_info.csv')
df
```

```python
df['team'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
df['team'].value_counts().plot(kind='bar')
plt.title('Number of Players in Each Team')
plt.xlabel('Team')
plt.ylabel('Players')
plt.show()
```

```python
df['playingRole'].value_counts()
df['playingRole'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Playing Role Distribution')
plt.show()
```

```python
df['battingStyle'].value_counts()
df['battingStyle'].value_counts().plot(kind='bar')
plt.title('Batting Style Distribution')
plt.show()
```

```python
df['bowlingStyle'].value_counts().head(10)
df['bowlingStyle'].value_counts().head(10).plot(kind='bar')
plt.title('Top Bowling Styles')
plt.show()

```

```python
pd.crosstab(df['team'], df['playingRole'])
```

```python
import seaborn as sns

role_team = pd.crosstab(df['team'], df['playingRole'])

sns.heatmap(role_team, annot=True, cmap='Blues')
plt.show()
```

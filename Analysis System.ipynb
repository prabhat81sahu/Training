```python
import numpy as np
import pandas as pd
```

```python
stud = pd.read_csv('/content/StudentsPerformance.csv')
stud.head(10)
```

```python
stud['overall_average_score'] = stud[['math score', 'reading score', 'writing score']].mean(axis=1)
display(stud.head())
```

```python
print(f"Overall average score of all students: {stud['overall_average_score'].mean():.2f}")
```

### Top 10 Performers

```python
top_10_performers = stud.sort_values(by='overall_average_score', ascending=False).head(10)
display(top_10_performers)
```

### Subject-wise Highest Scores

```python
highest_math = stud.loc[stud['math score'].idxmax()]
highest_reading = stud.loc[stud['reading score'].idxmax()]
highest_writing = stud.loc[stud['writing score'].idxmax()]

print(f"Highest Math Score: {highest_math['math score']} by student (index: {highest_math.name})")
print(f"Highest Reading Score: {highest_reading['reading score']} by student (index: {highest_reading.name})")
print(f"Highest Writing Score: {highest_writing['writing score']} by student (index: {highest_writing.name})")
```

### Assigning Grades

```python
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

stud['grade'] = stud['overall_average_score'].apply(assign_grade)
display(stud.head())
```

### Summary Statistics

```python
display(stud[['math score', 'reading score', 'writing score', 'overall_average_score']].describe())
```

```python
display(stud['grade'].value_counts().sort_index())
```

```python
emp = pd.read_excel('/content/employee_data.xlsx')
emp.head(10)
```

### Department-wise Average Salary

```python
def assign_department(age):
    if age < 30:
        return 'Marketing'
    elif 30 <= age < 40:
        return 'Sales'
    else:
        return 'HR'

emp['Department'] = emp['Age'].apply(assign_department)
display(emp.head())
```

```python
department_avg_salary = emp.groupby('Department')['Salary'].mean().reset_index()
display(department_avg_salary)
```

### Highest and Lowest Paid Employees

```python
highest_paid_employee = emp.loc[emp['Salary'].idxmax()]
lowest_paid_employee = emp.loc[emp['Salary'].idxmin()]

print("Highest Paid Employee:")
display(highest_paid_employee)

print("\nLowest Paid Employee:")
display(lowest_paid_employee)
```

### Compare Salaries by Gender

```python
gender_avg_salary = emp.groupby('Gender')['Salary'].mean().reset_index()
display(gender_avg_salary)
```

### Analyze Salary Trends

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(emp['Salary'], kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# Salary vs Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=emp, hue='Gender')
plt.title('Salary vs Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

### Generate Salary Reports

```python
# Overall Salary Report
print("Overall Salary Statistics:")
display(emp['Salary'].describe())

# Department-wise Salary Report
print("\nDepartment-wise Salary Statistics:")
display(emp.groupby('Department')['Salary'].describe())

# Gender-wise Salary Report
print("\nGender-wise Salary Statistics:")
display(emp.groupby('Gender')['Salary'].describe())
```
